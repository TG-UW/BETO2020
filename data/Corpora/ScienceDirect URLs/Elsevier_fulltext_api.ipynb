```python
df1 = pd.read_excel('flame_retardant_splits.xlsx','Sheet1')
df2 = pd.read_excel('flame_retardant_splits.xlsx','Sheet2')
df3 = pd.read_excel('flame_retardant_splits.xlsx','Sheet3')
df4 = pd.read_excel('flame_retardant_splits.xlsx','Sheet4')
```

```python
df2 = df2[['number','year','url','DJ score','WT score','AC score','JO score','AJ score','pdf only?']]
```

```python
df = df1
df = df.append(df2,ignore_index=True)
df = df.append(df3,ignore_index=True)
df = df.append(df4,ignore_index=True)

df = df.sort_values(by = 'number')
df = df.reset_index(drop=True)
```

```python
#This makes sure that all the links are of standard format, some links that get scraped don't have the 
#same format as most 
piis = []
for i in range(len(df)):
    pi = df.iloc[i,2][-17:]
    piis.append(pi)
```

```python
df['success'] = None
df.head()
```

```python
client = ElsClient('506e073160e8aa445febbadf41f11c83') # establish client 
```

```python
## ScienceDirect (full-text) document example using PII
def get_doc(dtype,identity):
    if dtype == 'pii':
        doc = FullDoc(sd_pii = identity)
    elif dtype == 'doi':
        doc= FullDoc(doi = identity)
       
    if doc.read(client):
            #print ("doc.title: ", doc.title)
            doc.write()   
    else:
        print ("Read document failed.")
    
    return doc
       
        
```

```python
file = open('flame_retardant_texts.txt','w')
```

```python
attempts = 0
successes = 0
for pii in piis:
    attempts += 1
    if attempts % 100 == 0:
        print('Attempts: ',attempts)
    
    try:
        doc = get_doc('pii',pii)
        try:
            text = doc.data['originalText']
            file.write(text)
            file.write('\n')
            successes += 1
            print('Successes: ',successes)
            df['success'][attempts-1] = 1
        except:
            df['success'][attempts-1] = 0
            pass
        
    except:
        df['success'][attempts-1] = 0

df.to_csv('flame_retardant_link_success.txt')
```
