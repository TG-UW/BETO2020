```python
def find_nth(haystack, needle, n):
    start = haystack.find(needle)
    while start >= 0 and n > 1:
        start = haystack.find(needle, start+len(needle))
        n -= 1
    return start
```

```python
def clean_paper(paper):
    if paper.count('Highlights') != 0:
        hind = paper.find('Highlights')
        paper = paper[hind + len('Highlights'):]
    elif paper.count('Abstract') != 0: 
        aind = paper.find('Abstract')
        paper = paper[aind + len('Abstract'):]
    elif paper.count('Introduction') != 0: 
        iind = find_nth(paper,'Introduction',2)
        paper = paper[iind + len('Introduction'):]
    else:
        pass
    
    rind = paper.rfind('References')
    paper = paper[:rind]
              
    return paper

        
```

```python
for i in range(len(df)):
    paper = df.iloc[i,0]
    df.iloc[i,0] = clean_paper(paper)
```
